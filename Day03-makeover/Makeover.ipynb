```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos
train_processed = pd.read_csv("https://raw.githubusercontent.com/Gianmarcos97/30daychartchallenge/main/Day03-makeover/denguecompetencia.csv", delimiter=",")

# Seleccionar solo las columnas específicas
selected_columns = ['ndvi_ne', 'ndvi_se', 'ndvi_sw', 'precipitation_amt_mm', 'reanalysis_air_temp_k',
                    'reanalysis_dew_point_temp_k', 'reanalysis_max_air_temp_k', 'reanalysis_min_air_temp_k',
                    'reanalysis_precip_amt_kg_per_m2', 'reanalysis_relative_humidity_percent',
                    'station_avg_temp_c', 'station_max_temp_c', 'station_min_temp_c', 'station_precip_mm']
train_processed_selected = train_processed[selected_columns]

# Calcular la matriz de correlación
correlation_matrix = train_processed_selected.corr()

cmap = sns.diverging_palette(220, 20, sep=20, as_cmap=True)

# Configurar la visualización
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap=cmap, fmt=".2f", cbar=True)
plt.title('Matriz de correlación - DengAI: Predicting Disease Spread')
plt.xticks(rotation=45, ha='right')
plt.show()

```
