```python
import pandas as pd
from pywaffle import Waffle
import matplotlib.pyplot as plt

# Cargar los datos
data = pd.read_csv("https://raw.githubusercontent.com/Gianmarcos97/30daychartchallenge/main/Day08-circular/paises-con-mas-aportes.csv", delimiter=",")

# Ordenar los datos por número de documentos y seleccionar solo el top 5
data_sorted = data.sort_values(by='Documents', ascending=False).head(5)

# Crear una gráfica de waffle
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=20,
    values=data_sorted['Documents'],
    labels=list(data_sorted['Country']),
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    figsize=(12, 8)  # Ajusta el tamaño de la imagen aquí
)

# Agregar texto con la cantidad de aportes abajo de la gráfica
total_aportes = data_sorted['Documents'].sum()
plt.figtext(0.5, 0.26, f'Total de aportes: {total_aportes}', ha='center', fontsize=12)

plt.title('Top 5 Países con Más Aportes a la IA en 2020', fontweight='bold')
plt.show()
```
